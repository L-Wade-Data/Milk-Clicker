```python
### WITH Cow

import tkinter as tk
from tkinter import messagebox
from tkinter import PhotoImage
import threading
import time

class MilkClicker(tk.Tk): #Inherit tk.TK 
    def __init__(self): 
        ''' Set the initial state of MilkClicker and define all attributes '''
        super().__init__() # Invokes the constructor of the superclass tk.TK
        self.title("Milk Clicker")
        self.geometry("600x600")
        
        # Load Images
        self.bg_image = PhotoImage(file="farm_pic_background_600_600.png")
        self.milk_image = PhotoImage(file="cow_pic_100x100.png")
        
        # Setting the Background image
        self.bg_label = tk.Label(self, image=self.bg_image)
        self.bg_label.place(x=0, y=0, relwidth=1, relheight=1)

        ### Initialising Game Parameters
        self.milk_count = 0 # Starting with 0 milk
        
        # Cow
        self.milk_per_click = 1  # Initial milk generated per click
        self.cow_price = 50  # Starting price for buying another cow
        self.num_cows = 1  # Starting with one cow
        
        # Milk Maid
        self.milk_maid_price = 10 # Starting price of milk maid
        self.milk_maid_production = 0.1  # milk per second from milk maid
        self.milk_maid = 0 # Starting with 0 milk maids
        
        # Achievements
        self.achievements = [10, 100, 1000, 10000] # achievement thresholds
        self.achieved = [] # list of achieved achievements

        # Milk Generating Button
        self.milk_btn = tk.Button(self, image=self.milk_image, borderwidth=0, command=self.click_milk, relief = 'raised') # Each time button is clicked, click_milk is run
        self.milk_btn.place(relx=0.5, rely=0.5, anchor=tk.CENTER) # Position

        ### Labels
        # Milk Count Label
        self.label = tk.Label(self, text="Milk: 0", bg='#A1DBCD', fg='black', relief=tk.RAISED) 
        self.label.place(relx=0.1, rely=0.05) # Position

        # Cow Count Display
        self.cow_count_display = tk.Label(self, text=f"Cows: {self.num_cows} ({self.milk_per_click}l/click)", 
                                          bg='#A1DBCD', fg='black',relief=tk.RAISED)
        self.cow_count_display.place(relx=0.1, rely=0.1) # Position

        # Milk Maid Count Display
        self.milk_maid_count_display = tk.Label(self, text=f"Milk Maids: {self.milk_maid}", bg='#A1DBCD', fg='black', relief=tk.RAISED)
        self.milk_maid_count_display.place(relx=0.1, rely=0.15)  # Position

        ### Store
        # Store Title
        self.store_title = tk.Label(self, text="Store:", bg='#A1DBCD', fg='black', width=12, height=2, 
                                    relief = tk.FLAT, padx=3)
        self.store_title.place(relx=0.75, rely=0.05)  # Position


        # Buy Milk Maid Button
        self.buy_milk_maid_btn = tk.Button(self, command=self.buy_milk_maid, # Run buy_click_maid when button is pressed
                                           text=f"Buy a Milk Maid\nCost: {self.milk_maid_price}l milk", 
                                           bg='#A1DBCD', width=12)
        self.buy_milk_maid_btn.place(relx=0.75, rely=0.11)  # Position
        
        # Buy Another Cow Button
        self.buy_cow_btn = tk.Button(self, text=f"Buy a Cow\nCost: {self.cow_price}l milk", command=self.buy_another_cow, 
                                     bg='#A1DBCD', width=12)
        self.buy_cow_btn.place(relx=0.75, rely=0.17) # Position

        # Start the automatic milk generation
        self.start_milk_maid()

        ### Welcome Tutorial Message
        # Display welcome message and instructions
        self.after(100, self.show_welcome_message) # delay required to keep it within the main thread
 
        # Help Button
        self.help_btn = tk.Button(self, text="Help", command=self.show_welcome_message)
        self.help_btn.place(relx=0.94, rely=0)


    def show_welcome_message(self):
        instructions = """
        Welcome to Milk Clicker!

        Objective: Click on the cow to collect milk. Use the milk to buy milk maids who will generate milk automatically.

        Instructions:
        - Click the cow image to earn milk.
        - Accumulate enough milk to buy milk maids from the store.
        - Milk maids increase your milk production automatically.
        - Reach milk milestones to unlock achievements.

        Have fun!
        """
        messagebox.showinfo("Welcome to Milk Clicker!", instructions)

    
    def click_milk(self):
        '''Increment the milk count and update the label.'''
        self.milk_count += self.milk_per_click # milk gained per click dependent on number of cows owned
        self.update_milk_count()
        self.check_achievements()

    
    def update_milk_count(self):
        '''Update the displayed milk count.'''
        self.label.config(text=f"Milk: {self.milk_count:.1f}l")

    
    def check_achievements(self):
        '''Check and display achievements when milk accumulation thresholds are reached.'''
        for amount in self.achievements:
            if self.milk_count >= amount and amount not in self.achieved:
                messagebox.showinfo("Achievement Unlocked!", f"You have accumulated {amount} milk!")
                self.achieved.append(amount)

    
    def buy_milk_maid(self):
        '''Buy a milk maid if enough milk is available.'''
        if self.milk_count >= self.milk_maid_price: # Check if you have enough milk to buy milk maid
            self.milk_count -= self.milk_maid_price # Take milk_maid_price off of milk_count
            self.update_milk_count() # Update milk_count label
            self.milk_maid += 1 # Add one milk maid 
            self.milk_maid_price += self.milk_maid_price * 0.1  # Increase the cost for the next milk maid by 10%
            self.buy_milk_maid_btn.config(text=f"Buy a Milk Maid\nCost: {self.milk_maid_price:.1f} milk") # Update milk maid price button
            self.milk_maid_count_display.config(text=f"Milk Maids: {self.milk_maid} ({(self.milk_maid * self.milk_maid_production):.1f} l/sec)") # Update the Milk Maid Count Label
        else:
            missing = self.milk_maid_price - self.milk_count
            messagebox.showinfo("You don't have enough milk!", f"You need {missing:.1f} more milk!")


    def buy_another_cow(self):
        '''Buy another cow if enough milk is available, increasing milk per click.'''
        if self.milk_count >= self.cow_price:
            self.milk_count -= self.cow_price
            self.update_milk_count()
            self.num_cows += 1
            self.milk_per_click += 1  # Increase milk generated per click
            self.cow_price += self.cow_price * 0.5  # Increase the cost for the next cow by 50%
            self.buy_cow_btn.config(text=f"Buy a Cow\nCost: {self.cow_price:.1f} milk")
            self.cow_count_display.config(text=f"Cows: {self.num_cows}  ({self.milk_per_click}l/click)")
        else:
            missing = self.cow_price - self.milk_count
            messagebox.showinfo("Not Enough Milk!", f"You need {missing:.1f} more milk to buy a cow!")

    
    def start_milk_maid(self):
        '''Start a thread to generate milk automatically.'''
        def generate_milk():
            while True:
                self.milk_count += self.milk_maid_production * self.milk_maid # Increment milk count by production that second
                self.update_milk_count()
                #self.check_achievements() ## Currently x2 achievements...
                time.sleep(1)  # Update every second

        # Threading to allow milk count to increase in the background concurrently with button and other functionality
        thread = threading.Thread(target=generate_milk) # Separate thread for generate_milk
        thread.daemon = True  # Thread automatically closes when Milk Clicker is closed
        thread.start()


if __name__ == '__main__':
    game = MilkClicker()
    game.mainloop()
```
